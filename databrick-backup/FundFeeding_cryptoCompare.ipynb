```
from datetime import datetime
import requests
from pyspark.sql import SparkSession
import pyspark.pandas as ps
```

```
list_cryptos=['ETH','BTC','BNB','MATIC']
```

```
cryptocompare_key='032e063f0e294dc37e419ad00331a788bde34dcba70d27cd5e41bf2b07a88fc4'
headers = {'Authorization': 'Bearer ' + cryptocompare_key}
currency_pair='USD'
cols_crypto_ohlc=["datetime","open","high","low","close",]
n_days=5
```

```
for symbol  in list_cryptos:
  try:
    
    url = f"https://min-api.cryptocompare.com/data/v2/histoday?fsym={symbol}&tsym={currency_pair}&limit={n_days}"
    json_ohlc=requests.get(url, headers=headers).json()

    df_ohlc=ps.DataFrame(json_ohlc['Data']['Data'])
    df_ohlc=df_ohlc.query('close>0')
    df_ohlc['datetime'] = df_ohlc['time'].apply(lambda d:datetime.fromtimestamp(d) )
    df_ohlc=df_ohlc[cols_crypto_ohlc]
    print(df_ohlc)
    
  except Exception as ex:
     print(f'{symbol}: {ex}')

  
    
```
